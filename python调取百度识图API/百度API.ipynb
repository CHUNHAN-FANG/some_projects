### 生成图片地址集
本地已保存图像样本的情况下,生成图像文件地址,待识别用

* os.walk()<br>
模块os中的walk()函数可以遍历文件夹下所有的文件。
* os.listdir()<br>
用于返回指定的文件夹下包含的文件或文件夹名字的列表，这个列表按字母顺序排序。<br>
os.listdir()函数得到的是仅当前路径下的文件名，不包括子目录中的文件，所有需要使用递归的方法得到全部文件名

```python
import os
import numpy as np

file_path='.\Dpic\\'
picid = os.listdir(file_path)  #本地存储旅游图片的相对路径

# 获取所有图像文件名称picid
def text_save(filename, data):#filename为写入CSV文件的路径，data为要写入数据列表.
    file = open(filename,'a')
    for i in range(len(data)):
        s = str(data[i]).replace('[','').replace(']','')#去除[],这两行按数据不同，可以选择
        s = s.replace("'",'').replace(',','') +'\n'  #去除单引号，逗号，每行末尾追加换行符
        file.write(s)
    file.close()
    print("保存文件成功") 

text_save('picid.txt',picid)
```

```python
#逐行读txt文件
with open("picid.txt") as txt:
    content = txt.readlines() #读全部行
    txt.close()
lines =np.array(content) #转换成array 类型
num_of_instances = lines.size #整个txt的行数
print("number of instances: ", num_of_instances)

# 统一每行加字符串
list=[]
for i in  range(0,num_of_instances):
    name,label=lines[i].split(".") #以 .分割文件名字
    list.append(file_path+name+"."+label)#统一每行加字符串 相对路径 file_path
with open("image_path.txt", 'w')as F:
    F.writelines(list)#写入到另一个txt文件中
    F.close()
```

```python
#读取图片
def get_file_content(filePath):
    with open(filePath, 'rb') as fp:
        return fp.read()
```
